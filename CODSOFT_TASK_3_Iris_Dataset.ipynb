# **IRIS FLOWER CLASSIFICATION**

Apply GaussianNB on iris dataset to develop a model that can classify iris
flowers into different species based on their sepal and petal
measurements.

```python
#import libraries
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
# Load the Iris dataset
df = load_iris()
```

```python
# To view the data Convert it to a pandas DataFrame
data = pd.DataFrame(data=df.data, columns=df.feature_names)
```

```python
# Add the target column to the DataFrame
data['species'] = df.target
data
```

Extract features (X) and target labels (y) for classification

```python
X = df.data
```

```python
X.shape
```

```python
y = df.target
```

```python
y.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.25,random_state = 1)
```

```python
# Create an instance of the Gaussian Naive Bayes classifier
model = GaussianNB()
```

```python
# Fit the model to the training data
model.fit(xtrain,ytrain)
```

```python
# Predict the species for the test set
predictions = model.predict(xtest)
```

```python
# Import metrics for evaluating the model's performance
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Evaluate the model's accuracy
accuracy_score(ytest,predictions)
```

```python
# Generate a confusion matrix to evaluate the performance
confusion_matrix(ytest,predictions)
```

```python
# Define the function for user input and classification
def classify_iris():
    print("Enter the sepal length (cm):")
    sepal_length = float(input())

    print("Enter the sepal width (cm):")
    sepal_width = float(input())

    print("Enter the petal length (cm):")
    petal_length = float(input())

    print("Enter the petal width (cm):")
    petal_width = float(input())

    # Create the user input as a NumPy array
    user_input = np.array([[sepal_length, sepal_width, petal_length, petal_width]])

    # Predict the species
    prediction = model.predict(user_input)

    # Get the species name from the prediction
    species = df.target_names[prediction][0]

    print(f"The predicted Iris species is: {species}")

# Call the function to classify Iris based on user input
classify_iris()
```
